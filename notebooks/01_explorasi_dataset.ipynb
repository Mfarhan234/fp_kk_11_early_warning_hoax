```python
# nilai threshold yang kita pakai (boleh kamu tweak dikit)
KR_HOAX_MIN = 0.59
KR_VALID_MAX = 0.43

KF_HOAX_MIN = 0.44
KF_VALID_MAX = 0.32

def label_core(row):
    if row["label_raw"] == "hoax":
        if row["kredibilitas_rendah"] >= KR_HOAX_MIN and row["kecurigaan_format"] >= KF_HOAX_MIN:
            return "core_hoax"
        else:
            return "ambiguous_hoax"
    elif row["label_raw"] == "valid":
        if row["kredibilitas_rendah"] <= KR_VALID_MAX and row["kecurigaan_format"] <= KF_VALID_MAX:
            return "core_valid"
        else:
            return "ambiguous_valid"
    else:
        return "other"

df_sample_100_llm["core_flag"] = df_sample_100_llm.apply(label_core, axis=1)
df_sample_100_llm["core_flag"].value_counts()
```

```python
# Pisah per kategori
df_amb_hoax  = df_sample_100_llm[df_sample_100_llm["core_flag"] == "ambiguous_hoax"].copy()
df_amb_valid = df_sample_100_llm[df_sample_100_llm["core_flag"] == "ambiguous_valid"].copy()

df_amb_hoax.shape, df_amb_valid.shape
```

```python
# 5 contoh ambiguous_hoax (random)
df_amb_hoax_sample = df_amb_hoax.sample(5, random_state=1)[
    ["sample_id", "text", "label_raw",
     "intensitas_emosi", "kecurigaan_format", "kredibilitas_rendah"]
]
df_amb_hoax_sample
```

```python
# 5 contoh ambiguous_valid (random)
df_amb_valid_sample = df_amb_valid.sample(5, random_state=2)[
    ["sample_id", "text", "label_raw",
     "intensitas_emosi", "kecurigaan_format", "kredibilitas_rendah"]
]
df_amb_valid_sample
```

```python
KR_HOAX_MIN = 0.59
KR_VALID_MAX = 0.43

KF_HOAX_MIN = 0.44
KF_VALID_MAX = 0.32

def explain_ambiguity(row):
    kr = row["kredibilitas_rendah"]
    kf = row["kecurigaan_format"]
    if row["label_raw"] == "hoax":
        reasons = []
        if kr < KR_HOAX_MIN:
            reasons.append(f"kredibilitas_rendah {kr:.2f} < batas hoax {KR_HOAX_MIN}")
        if kf < KF_HOAX_MIN:
            reasons.append(f"kecurigaan_format {kf:.2f} < batas hoax {KF_HOAX_MIN}")
        return "; ".join(reasons)
    elif row["label_raw"] == "valid":
        reasons = []
        if kr > KR_VALID_MAX:
            reasons.append(f"kredibilitas_rendah {kr:.2f} > batas valid {KR_VALID_MAX}")
        if kf > KF_VALID_MAX:
            reasons.append(f"kecurigaan_format {kf:.2f} > batas valid {KF_VALID_MAX}")
        return "; ".join(reasons)
    else:
        return "-"

# Contoh penjelasan untuk 5 ambiguous hoax
df_amb_hoax_sample.assign(alasan=lambda d: d.apply(explain_ambiguity, axis=1))
```

## Bagian C — PSO dengan 62 data core (UMPO sample)


```python
import numpy as np
from src.pso_optimizer import optimize_fuzzy_scaling_with_pso
from src.fuzzy_system import hitung_resiko_hoaks

cols_skor = ["intensitas_emosi", "kecurigaan_format", "kredibilitas_rendah"]

# 1. Ambil hanya core_hoax & core_valid
df_core = df_sample_100_llm[
    df_sample_100_llm["core_flag"].isin(["core_hoax", "core_valid"])
].copy()

# label target 0 / 100
df_core["label_resiko_hoaks"] = np.where(
    df_core["label_raw"] == "hoax", 100.0, 0.0
)

df_core["core_flag"].value_counts(), df_core.shape
```

```python
df_core.head()
```

```python
# Label target resiko hoaks (0–100)
df_core["label_resiko_hoaks"] = np.where(
    df_core["label_raw"] == "hoax", 100.0, 0.0
)
```

```python
from src.pso_optimizer import optimize_fuzzy_scaling_with_pso

result_pso = optimize_fuzzy_scaling_with_pso(
    df=df_core,
    target_col="label_resiko_hoaks",
    n_particles=40,        # boleh 30 dulu kalau mau cepat
    n_iters=80,            # bisa kamu kecilkan untuk tes awal, misal 40
    bounds_scale=(0.5, 1.5),
    seed=42,
)

result_pso
```

```python
best_w = result_pso["best_position"]
best_mse = result_pso["best_fitness"]

print("w terbaik  :", best_w)   # [w_emosi, w_format, w_kred]
print("MSE terbaik:", best_mse)
```

```python
from src.fuzzy_system import hitung_resiko_hoaks

def pred_resiko_with_w(df, w):
    emo  = np.clip(df["intensitas_emosi"].to_numpy() * w[0], 0.0, 1.0)
    frm  = np.clip(df["kecurigaan_format"].to_numpy() * w[1], 0.0, 1.0)
    kred = np.clip(df["kredibilitas_rendah"].to_numpy() * w[2], 0.0, 1.0)

    preds = [
        hitung_resiko_hoaks(e, f, k, output_scale_100=True)
        for e, f, k in zip(emo, frm, kred)
    ]
    return np.array(preds)

y_true = df_core["label_resiko_hoaks"].to_numpy()
y_pred = pred_resiko_with_w(df_core, best_w)

mse  = np.mean((y_pred - y_true)**2)
rmse = np.sqrt(mse)
print("RMSE di core set:", rmse)

# Kalau mau lihat akurasi klasifikasi (threshold 50)
y_true_cls = (y_true >= 50).astype(int)
y_pred_cls = (y_pred >= 50).astype(int)
acc = (y_true_cls == y_pred_cls).mean()
print("Akurasi klasifikasi (threshold 50):", acc)
```

```python
# pakai df_sample_100_llm utuh
y_true_full = np.where(df_sample_100_llm["label_raw"] == "hoax", 100.0, 0.0)
y_pred_full = pred_resiko_with_w(df_sample_100_llm, best_w)

mse_full = np.mean((y_pred_full - y_true_full)**2)
rmse_full = np.sqrt(mse_full)
print("RMSE di 100 sampel (core + ambiguous):", rmse_full)

y_true_full_cls = (y_true_full >= 50).astype(int)
y_pred_full_cls = (y_pred_full >= 50).astype(int)
acc_full = (y_true_full_cls == y_pred_full_cls).mean()
print("Akurasi di 100 sampel:", acc_full)
```

```python
df_amb = df_sample_100_llm[
    df_sample_100_llm["core_flag"].isin(["ambiguous_hoax", "ambiguous_valid"])
].copy()

y_true_amb = np.where(df_amb["label_raw"] == "hoax", 100.0, 0.0)
y_pred_amb = pred_resiko_with_w(df_amb, best_w)

mse_amb = np.mean((y_pred_amb - y_true_amb)**2)
rmse_amb = np.sqrt(mse_amb)
print("RMSE di ambiguous set:", rmse_amb)

y_true_amb_cls = (y_true_amb >= 50).astype(int)
y_pred_amb_cls = (y_pred_amb >= 50).astype(int)
acc_amb = (y_true_amb_cls == y_pred_amb_cls).mean()
print("Akurasi di ambiguous set:", acc_amb)
```

```python
df_amb = df_sample_100_llm[
    df_sample_100_llm["core_flag"].isin(["ambiguous_hoax", "ambiguous_valid"])
].copy()

df_amb["pred_resiko"] = pred_resiko_with_w(df_amb, best_w)

df_amb.groupby(["core_flag", "label_raw"])[
    ["intensitas_emosi", "kecurigaan_format", "kredibilitas_rendah", "pred_resiko"]
].mean()
```
